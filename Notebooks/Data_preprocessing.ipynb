'Andhra Pradesh' , 'Andhra Pradesh '              
'Assam' , 'Assam '            
'Bihar' , 'Bihar '.....

2. Inconsistent : 'and' and '&' 

```python
# trim space nd ... 
Data_rape['State/UT'] = Data_rape['State/UT'].str.strip()
```

```python
Data_rape['State/UT'] = Data_rape['State/UT'].str.replace('&', 'and')
```

```python
Data_rape['State/UT'] = Data_rape['State/UT'].str.title()
```

3. Improper State/UT names

```python
Data_rape['State/UT'] = Data_rape['State/UT'].replace( 
    {
    'A And N Islands': 'Andaman And Nicobar Islands',
    'Delhi Ut': 'Delhi',
    'D And N Haveli': 'Dadra And Nagar Haveli And Daman And Diu',
    'Dandn Haveli': 'Dadra And Nagar Haveli And Daman And Diu',
    'Dadra And Nagar Haveli': 'Dadra And Nagar Haveli And Daman And Diu',
    'Daman And Diu': 'Dadra And Nagar Haveli And Daman And Diu',
    'D And N Haveli And Daman And Diu': 'Dadra And Nagar Haveli And Daman And Diu'
} )
```

```python
Data_rape['State/UT'].unique()
```

4. Founded 2 Invalid state names : 'Total State (S)' and 'Total State(S)' 


```python
Data_rape = Data_rape[~Data_rape['State/UT'].str.contains('Total', case=False)]
```

5. Meghalaya and meghalaya* both are diffrent 

```python
Data_rape['State/UT'] = Data_rape['State/UT'].replace({
    'Meghalaya*':'Meghalaya'
})
```
